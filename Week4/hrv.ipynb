```python
import heartpy as hp
import pandas as pd
import numpy as np
```

```python
col_names = ['I', 'II', 'III', 'aVR', 'aVL', 'aVF', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6']
df1 = pd.read_csv("Sample_SB.csv", names=col_names)
df2 = pd.read_csv("Sample_SR.csv", names=col_names)
df3 = pd.read_csv("Sample_ST.csv", names=col_names)
```

```python
df1.head()
```

```python
srate = 500
data1 = df1[col_names[0]]
wd, m = hp.process(data1, srate)
hp.plotter(wd, m)
```

```python
filtered = hp.filter_signal(data1, cutoff=0.05, sample_rate=srate, filtertype='notch')
wd, m = hp.process(hp.scale_data(filtered), srate)
```

```python
hp.plotter(wd, m)
```

```python
data2 = df2[col_names[0]]
wd, m = hp.process(data2, srate)
filtered = hp.filter_signal(data2, cutoff=0.05, sample_rate=srate, filtertype='notch')
wd, m = hp.process(hp.scale_data(filtered), srate)
hp.plotter(wd, m)
```

```python
data3 = df3[col_names[0]]
wd, m = hp.process(data3, srate)
filtered = hp.filter_signal(data3, cutoff=0.05, sample_rate=srate, filtertype='notch')
wd, m = hp.process(hp.scale_data(filtered), srate)
hp.plotter(wd, m)
```
